# Problem 2 Indexing, Subsetting and Slicing 
## This problem is seperated into 4 parts after accomplishing all of them it is then saved as a .py file with the name Elefante_Pandas-P2.py

## Problem 2a: The first part of problem 2 in which the task is to show the odd number columns of the first five rows and then displaying them. This was accomplished using the iloc function.

```python
#Problem 2
#Part a
#Use the iloc function to specify which columns and rows should be displayed
Cars.iloc[[0,1,2,3,4],[1,3,5,7,9,11]]
```

## Problem 2b: The second part of problem 2 in which the task is to show and display the row that has the model Mazda RX4 and was accomplished using slicing.

```python
#Problem 2
#Part b
#Slicing the data frame to only display the row of the Mazda RX4 which is row 0 and column 0
Cars[0:1]
```

## Problem 2c: The third part of problem 2 in which the task is to show and display the how many cylinders the model Camaro Z28 has. It was accomplished using the .loc function

```python
#Problem 2
#Part c
#Using the .loc function which acts similarly to a search engine to display how many cylinders the car Camaro Z28 has.
Cars.loc[(Cars['Model']=='Camaro Z28'), ['Model','cyl']]
```

## Problem 2d: The last part of problem 2 in which the task is to show and display the number of cylinders and gears that the cars Mazda RX4, Ford Pantera L, and the Honda Civic. It was accomplished using the both the concat and iloc functions.

```python
# Problem 2
#Part d
#Using the concat function in conjuction with the .loc function to display the how many cylinders and gears the Mazda RX4 Wag, Ford Pantera L and Honda Civic has.
pd.concat([Cars.loc[(Cars['Model']=='Mazda RX4 Wag'), ['Model','cyl','gear']], Cars.loc[(Cars['Model']=='Ford Pantera L'), ['Model','cyl','gear']], Cars.loc[(Cars['Model']=='Honda Civic'), ['Model','cyl','gear']]])
```
